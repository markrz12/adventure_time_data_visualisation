# Lumpy Space Theme

```python
colors = [(51/255, 1/255, 45/255), 
    (84/255, 0/255, 85/255), 
    (205/255, 165/255, 225/255), 
    (27/255, 197/255, 214/255), 
    (98/255, 121/255, 216/255)]

my_cmap = ListedColormap(colors, name='lumpy')
mpl.colormaps.register(cmap=my_cmap)

cmap2 = LinearSegmentedColormap.from_list("lumpy", colors)
mpl.colormaps.register(cmap=cmap2, name="lumpy_theme")
```

```python
plt.style.use("lumpy_space_theme")
```

```python
fruits = ['Apples', 'Bananas', 'Cherries']
group_a_votes = [150, 120, 90]
group_b_votes = [130, 140, 100]
group_c_votes = [160, 110, 95]
group_d_votes = [140, 130, 85]


x = np.arange(len(fruits))

width = 0.2

plt.bar(x - 1.5*width, group_a_votes, width, label='Group A')
plt.bar(x - 0.5*width, group_b_votes, width, label='Group B')
plt.bar(x + 0.5*width, group_c_votes, width, label='Group C')
plt.bar(x + 1.5*width, group_d_votes, width, label='Group D')


plt.xlabel('Fruit')
plt.ylabel('Number of Votes')
plt.title('Lumpy Space Theme')
plt.xticks(x, fruits)
plt.legend()

plt.show()
```

```python
data1_x = [1, 2, 3, 4, 5, 6, 7]
data1_y = [4, 3, 5, 4, 6, 5, 7]

data2_x = [2, 3, 4, 5, 6, 7, 8]
data2_y = [3, 4, 2, 5, 4, 6, 5]

data3_x = [1, 3, 4, 6, 7, 8, 9]
data3_y = [6, 5, 3, 4, 6, 7, 5]

data4_x = [2, 9,3, 5, 2, 1, 8]
data4_y = [8, 7, 5, 2, 5, 1, 2]

#plt.figure(figsize=(8,6))
plt.scatter(data1_x, data1_y, label='Dataset 1')
plt.scatter(data2_x, data2_y, label='Dataset 2')
plt.scatter(data3_x, data3_y, label='Dataset 3')
plt.scatter(data4_x, data4_y, label='Dataset 3')

plt.title("Lumpy Space Theme")
plt.legend()
plt.show()
```

```python
x = np.array([1, 2, 3, 4])  
y = x*2  

x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]
 
x2 = [3, 4, 5, 6, 7]
y2 = [2, 3, 4, 5, 6]

x3 = [5, 6, 7]
y3 = [3,2,1]

 
plt.plot(x, y, label = "Group A") 
plt.plot(x1, y1, label = "Group B") 
plt.plot(x2, y2, label = "Group C") 
plt.plot(x3, y3, label = "Group D") 


plt.title("Lumpy Space Theme")
plt.legend()
plt.show()
```

```python
data = np.random.rand(10, 10)
plt.imshow(data)
plt.title("Lumpy Space Theme")
plt.colorbar()
plt.show()
```
