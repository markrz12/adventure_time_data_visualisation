```python
import matplotlib.pyplot as plt
from cycler import cycler
import numpy as np
import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap, ListedColormap
import seaborn as sns
```

```python
print(mpl.get_configdir())
```

# Lumpy Space Theme

```python
colors = [(51/255, 1/255, 45/255), 
    (84/255, 0/255, 85/255), 
    (205/255, 165/255, 225/255), 
    (27/255, 197/255, 214/255), 
    (98/255, 121/255, 216/255)]

my_cmap = ListedColormap(colors, name='lumpy')
mpl.colormaps.register(cmap=my_cmap)

cmap2 = LinearSegmentedColormap.from_list("lumpy", colors)
mpl.colormaps.register(cmap=cmap2, name="lumpy_theme")
```

```python
plt.style.use("lumpy_space_theme")
```

```python
fruits = ['Apples', 'Bananas', 'Cherries']
group_a_votes = [150, 120, 90]
group_b_votes = [130, 140, 100]
group_c_votes = [160, 110, 95]
group_d_votes = [140, 130, 85]
x = np.arange(len(fruits))
width = 0.2

data1_x = [1, 2, 3, 4, 5, 6, 7]
data1_y = [4, 3, 5, 4, 6, 5, 7]
data2_x = [2, 3, 4, 5, 6, 7, 8]
data2_y = [3, 4, 2, 5, 4, 6, 5]
data3_x = [1, 3, 4, 6, 7, 8, 9]
data3_y = [6, 5, 3, 4, 6, 7, 5]
data4_x = [2, 9,3, 5, 2, 1, 8]
data4_y = [8, 7, 5, 2, 5, 1, 2]

x_line = np.array([1, 2, 3, 4])  
y_line = x_line * 2
x1_line = [2, 4, 6, 8]
y1_line = [3, 5, 7, 9]
x2_line = [3, 4, 5, 6, 7]
y2_line = [2, 3, 4, 5, 6]
x3_line = [5, 6, 7]
y3_line = [3, 2, 1]

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[1, 0].bar(x - 1.5*width, group_a_votes, width, label='Group A')
axs[1, 0].bar(x - 0.5*width, group_b_votes, width, label='Group B')
axs[1, 0].bar(x + 0.5*width, group_c_votes, width, label='Group C')
axs[1, 0].bar(x + 1.5*width, group_d_votes, width, label='Group D')
axs[1, 0].set_xlabel('Fruit')
axs[1, 0].set_ylabel('Number of Votes')
axs[1, 0].set_title(f' Bar Plot')
axs[1, 0].set_xticks(x)
axs[1, 0].set_xticklabels(fruits)
axs[1, 0].legend()

axs[0, 1].scatter(data1_x, data1_y, label='Group A')
axs[0, 1].scatter(data2_x, data2_y, label='Group B')
axs[0, 1].scatter(data3_x, data3_y, label='Group C')
axs[0, 1].scatter(data4_x, data4_y, label='Group D')
axs[0, 1].set_title(f'Scatter Plot')
axs[0, 1].set_xlabel('Number of x')
axs[0, 1].set_ylabel('Number of y')


axs[0, 1].legend()

axs[0, 0].plot(x_line, y_line, label='Group A')
axs[0, 0].plot(x1_line, y1_line, label='Group B')
axs[0, 0].plot(x2_line, y2_line, label='Group C')
axs[0, 0].plot(x3_line, y3_line, label='Group D')
axs[0, 0].set_title(f'Line Plot')
axs[0, 0].set_xlabel('Number of x')
axs[0, 0].set_ylabel('Number of y')
axs[0, 0].legend()

data = np.random.rand(10, 10)
im = axs[1, 1].imshow(data)
fig.colorbar(im, ax=axs[1, 1])
axs[1, 1].set_title(f'Image Plot')
axs[1, 1].grid(False)

plt.show()
```
