```python
import matplotlib.pyplot as plt
from cycler import cycler
import numpy as np
import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap, ListedColormap
```

```python
colors = [
    (51/255, 1/255, 45/255), 
    (84/255, 0/255, 85/255), 
    (205/255, 165/255, 225/255), 
    (27/255, 197/255, 214/255), 
    (98/255, 121/255, 216/255), 
]


my_cmap = ListedColormap(colors, name='lumpy2')

mpl.colormaps.register(cmap=my_cmap)

cmap2 = LinearSegmentedColormap.from_list("lumpy2", colors)

mpl.colormaps.register(cmap=cmap2, name="lumpy_theme2")
```

```python
def lumpy_space():
    plt.style.use('default')
    plt.rcParams.update({
        'axes.prop_cycle': cycler('color', ['#c085dd', '#35accf', '#636aba', '#415199']),
        'axes.facecolor': '#160e23',
        'axes.edgecolor': '#c085dd',
        'figure.facecolor': '#461854',
        'text.color': '#c085dd',
        'xtick.color': '#c085dd',
        'ytick.color': '#c085dd',
        'xtick.direction': 'in',          
        'ytick.direction': 'in',         
        'font.family': 'Comic Sans MS',
        'font.size': 12,
        'patch.edgecolor': '#9C84A6',
        'patch.linewidth': 1.5,
        'legend.facecolor': '#461854',
        'legend.edgecolor': '#c085dd',
        'legend.fontsize': 12,
        'axes.titlesize': 18,
        'axes.titlecolor': '#c085dd',     
        'axes.labelsize': 14,
        'axes.labelcolor': '#c085dd',    
        'axes.titlepad': 12,
        'figure.autolayout': True,     
        'scatter.marker': 'D',
        'lines.markersize': 9,
        'lines.linewidth': 0.75,
        'lines.marker': 'D',
        'lines.linestyle': '-',
        'image.cmap': 'lumpy_theme2',
        'axes.grid': True,
        'axes.grid.which': 'both',
        'grid.alpha': 0.3,
        'axes.axisbelow': True
    })
```

```python
lumpy_space()

fruits = ['Apples', 'Bananas', 'Cherries']
group_a_votes = [150, 120, 90]
group_b_votes = [130, 140, 100]
group_c_votes = [160, 110, 95]
group_d_votes = [140, 130, 85]


x = np.arange(len(fruits))

width = 0.2

#plt.figure(figsize=(8, 6))
plt.bar(x - 1.5*width, group_a_votes, width, label='Group A')
plt.bar(x - 0.5*width, group_b_votes, width, label='Group B')
plt.bar(x + 0.5*width, group_c_votes, width, label='Group C')
plt.bar(x + 1.5*width, group_d_votes, width, label='Group D')


plt.xlabel('Fruit')
plt.ylabel('Number of Votes')
plt.title('Lumpy Space Theme')
plt.xticks(x, fruits)
plt.legend()

plt.show()
```
